```python
#Importar as bibliotecas necessárias
import pandas as pd

#Ler tabela
tabela = pd.read_excel('banco_2025.xlsx')

#Exibir tabela
display(tabela)

#Exibir informações da tabela - nomes das colunas e tipos de dados
display(tabela.info())

#Verificar valores nulos na tabela
display(tabela.isnull().sum())
```

```python
#Deletar coluna "data_referência" -> Todas as linhas se referem ao mesmo dia: 01/01/2025
tabela = tabela.drop('data_referencia', axis=1)

#Deletar a coluna 'nao_informado' -> Fizemos isso pois essa coluna não tem relevância para o nosso projeto
tabela = tabela.drop('nao_informado', axis=1)

#Coluna "total" -> Não sabemos ao quê se refere
tabela = tabela.drop('total', axis=1)

#Deletar coluna "abrangencia" -> Todas as linhas dessa coluna estão preenxidas com a mesma palavra "Estadual"
tabela = tabela.drop('abrangencia', axis=1)

#Deletar a coluna "formulario" -> não sabemos sobre o que se trata e não parece ter relevância para a nossa análise
tabela = tabela.drop('formulario', axis=1)

display(tabela)
```

```python
#Substituindo valores vazios

#Agente
tabela['agente'] = tabela['agente'].fillna('Não informado')

#Arma
tabela['arma'] = tabela['arma'].fillna('Não Informado')

#Faixa etária
tabela['faixa_etaria'] = tabela['faixa_etaria'].fillna('Idade Não Informada')

#Feminino
tabela['feminino'] = tabela['feminino'].fillna(0).astype(int)

#Masculino
tabela['masculino'] = tabela['masculino'].fillna(0).astype(int)

#Total vítimas
tabela['total_vitima'] = tabela['total_vitima'].fillna(0).astype(int)

#Total peso
tabela['total_peso'] = tabela['total_peso'].fillna(0)

#Mostrar as alterações na tabela:
display(tabela)
display(tabela.info())
display(tabela.isnull().sum())
```

```python
#Criar um novo arquivo excel com as alterações feitas na tabela
tabela.to_excel('dados_tratados.xlsx')
```
